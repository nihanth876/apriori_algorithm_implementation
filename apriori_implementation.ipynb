```python
import collections
import itertools
import ast
def apriori_implementation(df1,test,threshold,method):
  min_sup = threshold*int(df1.shape[0])

  test = collections.OrderedDict(sorted(test.items()))
  all_elem = {}
  if method=='f1':    
    k = 1
    candidates = {}    
    while True:
      if k==1:
        final = {}
        count=0        
        candidates[str(k)] = len(test)
        for i in test.keys():          
          if int(test[i])>=min_sup:
            final[i]=test[i]
          else:
            continue
        final = collections.OrderedDict(sorted(final.items()))
        print('frequent 1 itemsets are: '+str(final.keys()))        
        print('frequent '+str(k)+' itemsets count is: '+str(len(final)))
        print('candidates counts is: '+str(candidates[str(k)]))
        print('\n')
        if len(final)==0:
          break
        else:
          all_elem = {'1':final}
          k=k+1

      elif k==2:        
        final = all_elem['1']
        temp = list(itertools.combinations(final.keys(),k))
        candidates[str(k)]=len(temp)        
        a = {}
        for i in temp:              
          i = sorted(list(i))          
          for j in range(5):
            elem = list(df1.iloc[j].fillna(''))            
            if set(elem)>=set(i):  #comparing both elements to check if they are subsets and thus present in the transaction to increase support count
              if a.get(str(i)):          
                a[str(i)]+=1
              else:
                a[str(i)]=1
        returned = {}
        for i in a.keys():
          if a[i]>=min_sup:
            returned[i]=a[i]        
        print('candidates counts for k=2 is: '+str(candidates[str(k)]))
        if len(returned.keys())==0:
          print('frequent 2 itemsets are: NA')
          print('\n')
          break
        else:
          print('frequent 2 itemsets are: '+str(returned.keys()))
          print('frequent '+str(k)+' itemsets count is: '+str(len(returned.keys())))
          print('\n')
          all_elem['2']=returned
          k=k+1
      else:
        next= []
        a={}
        b={}
        final = all_elem[str(k-1)]
        f1 = all_elem['1']
        for i in final.keys():
          temp = ast.literal_eval(i)          
          for j in f1.keys():
            if j in temp:
              continue
            else:
              temp1 = temp+[j]
              flag=0
              for mn in next:
                if sorted(mn)==sorted(temp1):
                  flag=1
                  break
              if flag==0:                
                next.append(sorted(temp1))   #appending sorted keys to make sure the lexicographic order is preserved 
        
        candidates[str(k)]=len(next)
        print('candidates counts for k='+str(k)+' is '+str(candidates[str(k)]))
        for i in next:
          flag=[]
          #flag1=0
          count_candidates = 0
          for y in range(k-2,0,-1):
            temp = list(itertools.combinations(i,y+1))            
            for kl in temp:
              for m in all_elem[str(y+1)].keys():   #changed y to y+1 as we are using 1 indexing
                temp2 = ast.literal_eval(m)         #to convert "[a,b,c]" to [a,b,c]                
                if sorted(temp2) == sorted(list(kl)):              
                  flag.append('1')
                  break
                else:              
                  continue          
            if len(flag)==len(temp):
              b[str(i)] = 0
          #print("b is:"+str(b))
        for i in b.keys():
            temp = ast.literal_eval(i)            
            for j in range(5):
              elem = list(df1.iloc[j].fillna(''))
              #print(elem)
              if len(elem)<len(temp):
                continue            
              if set(elem)>=set(temp):
                if a.get(str(temp)):
                  a[str(temp)]+=1
                else:
                  a[str(temp)]=1        
        candidates[str(k)]=len(a.keys())
        print('candidates counts for k='+str(k)+' after pruning is '+str(candidates[str(k)]))
        returned = {}
        for i in a.keys():
          if a[i]>=min_sup:
            returned[i]=a[i]
        if len(returned.keys())==0:
          print("No frequent itemsets for k="+str(k))
          break
        else:
          print('frequent '+str(k)+' itemsets are: '+str(returned.keys()))
          print('frequent '+str(k)+' itemsets count is: '+str(len(returned.keys())))
          print('\n')
          all_elem[str(k)]=returned
          k=k+1
  elif method=='fk_1':      
      k = 1
      candidates = {}
      #all_elem = {}
      while True:
        if k==1:
          final = {}
          count=0        
          candidates[str(k)] = len(test)
          for i in test.keys():        
            if int(test[i])>=min_sup:
              final[i]=test[i]
            else:
              continue
          final = collections.OrderedDict(sorted(final.items()))
          print('frequent 1 itemsets are: '+str(final.keys()))          
          print('frequent '+str(k)+' itemsets count is: '+str(len(final)))
          print('candidates counts is: '+str(candidates[str(k)]))
          print('\n')
          if len(final)==0:
            break
          else:
            all_elem = {'1':final}
            k=k+1

        elif k==2:        
          final = all_elem['1']
          temp = list(itertools.combinations(final.keys(),k))
          candidates[str(k)]=len(temp)
          #print(temp)
          a = {}
          for i in temp:    
            #i = list(i)  #was there before replacing with sorted(i)
            i = sorted(list(i))            
            for j in range(5):
              elem = list(df1.iloc[j].fillna(''))              
              if set(elem)>=set(i):
                if a.get(str(i)):          
                  a[str(i)]+=1
                else:
                  a[str(i)]=1
          returned = {}
          for i in a.keys():
            if a[i]>=min_sup:
              returned[i]=a[i]        
          print('candidates counts for k=2 is: '+str(candidates[str(k)]))
          if len(returned.keys())==0:
            print('frequent 2 itemsets are: NA')
            print('\n')
            break
          else:
            print('frequent 2 itemsets are: '+str(returned.keys()))
            print('frequent '+str(k)+' itemsets count is: '+str(len(returned.keys())))
            print('\n')
            all_elem['2']=returned
            k=k+1
        else:
          next= []
          a={}
          b={}
          final = all_elem[str(k-1)]   #getting the k-1 element to perform fk-1*fk-1
          f1 = all_elem[str(k-1)]          
          for i in final.keys():
            temp_fk_1 = ast.literal_eval(i)
            #print(type(temp_fk_1))
            for j in f1.keys():
              temp1_fk_1 = ast.literal_eval(j)
              if temp_fk_1==temp1_fk_1:
                continue
              else:
                #print(i)
                if temp_fk_1[0:k-2] == temp1_fk_1[0:k-2]:     #to make sure that the k-2 element is same so that we can combine both the k-1 elements
                  dummy = []
                  dummy.append(temp_fk_1[0:k-2])
                  temp_final_fk_1 = dummy[0]+[temp_fk_1[k-2]]+[temp1_fk_1[k-2]]
                  #print(temp_final_fk_1)
                  if sorted(temp_final_fk_1) not in next:
                    next.append(sorted(temp_final_fk_1))
                  
            
          candidates[str(k)]=len(next)
          #print('candidates for k='+str(k)+' after pruning is '+str(next))
          print('candidates counts for k='+str(k)+' is '+str(candidates[str(k)]))
          for i in next:
            flag=[]
            #flag1=0
            count_candidates = 0
            for y in range(k-2,0,-1):
              temp = list(itertools.combinations(i,y+1))            
              for kl in temp:
                for m in all_elem[str(y+1)].keys():   #changed y to y+1 as we are using 1 indexing
                  temp2 = ast.literal_eval(m)                  
                  if sorted(temp2) == sorted(list(kl)):              
                    flag.append('1')
                    break
                  else:              
                    continue          
              if len(flag)==len(temp):
                b[str(i)] = 0            
          for i in b.keys():    
              temp = ast.literal_eval(i)
              #print(temp)
              for j in range(5):
                elem = list(df1.iloc[j].fillna(''))
                #print(elem)
                if len(elem)<len(temp):
                  continue            
                if set(elem)>=set(temp):
                  if a.get(str(temp)):
                    a[str(temp)]+=1
                  else:
                    a[str(temp)]=1
          #print("a is: "+str(a))
          #candidates[str(k)]=len(next)
          candidates[str(k)]=len(a.keys())
          print('candidates counts for k='+str(k)+' after pruning is '+str(candidates[str(k)]))        
          returned = {}
          for i in a.keys():
            if a[i]>=min_sup:
              returned[i]=a[i]
          if len(returned.keys())==0:
            print("No frequent itemsets for k="+str(k))
            break
          else:
            print('frequent '+str(k)+' itemsets are: '+str(returned.keys()))
            print('frequent '+str(k)+' itemsets count is: '+str(len(returned.keys())))
            print('\n')
            all_elem[str(k)]=returned
            k=k+1
  else:
    print("invalid method name")
  return all_elem
```

```python
def rule_generation(y,min_conf):
  print('\n')
  u=1  
  final_rules = []
  for i in range(2,int(list(y.keys())[-1])+1):
    temp = y[str(i)]    
    z = []
    for j in temp.keys():
      temp1 = ast.literal_eval(j)
      z.append(temp1)    
    x = (set(list(itertools.chain.from_iterable(z))))    
    while int(i)>int(u):
      temp2 = list(itertools.combinations(x,u))
      conf=0 #was-1      
      for l in temp:
        temp3 = ast.literal_eval(l)        
        for k in temp2:
          support_mod = 0 
          support_fk = 0     
          if k[0] in temp3:            
            mod = [x for x in temp3 if x not in k]    #getting the array of elements fk - h(m +1) as in the textbook            
            if str(mod)+"->"+str(k[0]) in final_rules:
              continue
            support_fk = int(y[str(i)][str(temp3)])            
            try:              
              if u==1:                
                support_mod = int(y[str(u)][str(mod[0])])
              else:
                support_mod = int(y[str(u)][str(mod)])              
            except Exception as e:
              continue

            try:
              conf = support_fk/support_mod
            except Exception as e:
              continue            
            if conf>=min_conf:
              rule = str(mod)+"->"+str(k[0])
              final_rules.append(rule)
      u=u+1
  print("the final rules are:")
  print(final_rules)
  print('\n')
```

```python
#this is to generate the number of occurences of each item in the transactions to be passed to the apriori method
counts={}
for i in range(int(df.shape[0])):
      elem = df.iloc[i]
      num = elem['Item(s)']
      values = list(elem[1:num+1])
      for i in values:
        if counts.get(i):
          counts[i]+=1
        else:
          counts[i]=1
counts
```
